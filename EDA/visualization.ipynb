```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned datasets
trans = pd.read_csv('cleaned_trans.csv')
loan = pd.read_csv('cleaned_loan.csv')
```

```python
# Visualization 1: Distribution of Loan Amounts
plt.figure(figsize=(10, 6))
sns.histplot(loan['amount'], bins=30, kde=True)
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Daily Transaction Amounts Over Time
trans['date'] = pd.to_datetime(trans['date'], format='%Y-%m-%d')
```

```python
# Group by date
daily_transactions = trans.groupby(trans['date'].dt.date)['amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(daily_transactions['date'], daily_transactions['amount'], marker='o')
plt.title('Daily Transaction Amounts Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Visualization Completed!")
```
